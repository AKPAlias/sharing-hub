The aim of this project is to clean used car listings from eBay data and analyze results for useful insights.

- We have 20 columns in the dataset, of which 15 have no nulls (missing values). We have a few cells that can be converted to the numeric form including price and odometer. The notRepairedDamage column could be converted to boolean and the column names could be cleaned for easier reading.



```python
autos.columns
```

```python
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'abtest',
       'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
       'odometer', 'registration_month', 'fuel_type', 'brand',
       'unrepaired_damage', 'ad_created', 'nr_of_pictures', 'postal_code',
       'last_seen']
```

```python
autos.head()
```

- The column names used camelcase instead of Python's preferred snakecase. Changes to the column names have been made for readability of the user and computer.

```python
autos.describe(include="all")
```

- The nr_of_pictures column has only zeroes in it and so can be dropped as it is not useful. Also, the seller and offer_type are almost completely the same for each listing (private and Angebot, respectively) so these columns can be dropped as well.
- The maximum and minimum values for registration_year, power_ps seem to be weird and should be investigated for correct information and/or outliers.
- As noted earlier, the odometer and price columns have numeric data that should be converted to numeric.
- First we drop the three columns we deemed not useful.

```python
autos = autos.drop(['nr_of_pictures','seller','offer_type'],axis = 1)
```

```python
autos['odometer'] = (autos['odometer']
                         .str.replace(',','')
                         .str.replace('km','')
                         .astype(int)
                    )

autos['price'] = (autos['price']
                      .str.replace('$','')
                      .str.replace(',','')
                      .astype(int)
                  )

autos.rename(columns = {"odometer": "odometer_km"}, inplace=True)
autos.rename(columns = {"price": "price_dollars"}, inplace=True)
```

```python
autos['brand'].value_counts()
```

- The top five manufacturers on the list are all German brands with Volkswagen by far the most popular.
- We'll limit our aggregate analysis to brands in the top 20 car sales represented in our data set.

```python
brand_list = autos['brand'].value_counts()
brand_top_10 = brand_list[:10].index
brand_top_10
```

```python
brand_mean_prices = {}
for brand in brand_top_10:
    brand_mean_prices[brand] = autos[autos['brand'] == brand]['price_dollars'].mean()
    
brand_mean_prices
```

- Based on the results of our aggregrate data, it is clear that there are around  3 tiers interms of car sale prices.
- The first tier appears to be anything over \$8000 including Audi, BMW and Mercedes_Benz. This is to be expected as these are considered luxury car brands.
- The more common and affordable car brands such as Fiat, Ford, Opel, Peugeot, and Renault occupy the lower tier where cars cost anything lower than \$4000
- There also appears to be a middle tier around the \$5000 mark which is occupied by Volkswagen. Volkswagen has the largest number of data in our dataset, and it it can be assumed that this is a realistic result.
- Seat cars appear to be selling just around \$4250 which I would say is also close to this middle-tier and not a lower-tier car price as there is huge seperation between it and other lower-tier brands.

```python
brand_top_5 = brand_list[:5].index

brand_mean_prices = {}
for brand in brand_top_5:
    brand_mean_prices[brand] = autos[autos['brand'] == brand]['price_dollars'].mean()
    
brand_mean_mileage = {}
for brand in brand_top_5:
    brand_mean_mileage[brand] = autos[autos['brand'] == brand]['odometer_km'].mean()
```
